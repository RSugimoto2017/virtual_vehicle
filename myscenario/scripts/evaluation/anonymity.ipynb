```python
# ライブラリのインポート
import numpy as np
import pandas as pd
import os
```

```python
# 必要パラメータ
method = 'suggestion'
schedule = '8am'
change_pseudo_num = 10
threshold = 5
```

```python
anonymity_normal_3am_3 = np.ones(11)
anonymity_normal_3am_5 = np.ones(11)
anonymity_normal_3am_10 = np.ones(11)
anonymity_normal_8am_3 = np.ones(11)
anonymity_normal_8am_5 = np.ones(11)
anonymity_normal_8am_10 = np.ones(11)
entropy_normal_3am_3 = np.zeros(11)
entropy_normal_3am_5 = np.zeros(11)
entropy_normal_3am_10 = np.zeros(11)
entropy_normal_8am_3 = np.zeros(11)
entropy_normal_8am_5 = np.zeros(11)
entropy_normal_8am_10 = np.zeros(11)
anonymity_suggestion_3am_3 = np.ones(11)
anonymity_suggestion_3am_5 = np.ones(11)
anonymity_suggestion_3am_10 = np.ones(11)
anonymity_suggestion_8am_3 = np.ones(11)
anonymity_suggestion_8am_5 = np.ones(11)
anonymity_suggestion_8am_10 = np.ones(11)
entropy_suggestion_3am_3 = np.zeros(11)
entropy_suggestion_3am_5 = np.zeros(11)
entropy_suggestion_3am_10 = np.zeros(11)
entropy_suggestion_8am_3 = np.zeros(11)
entropy_suggestion_8am_5 = np.zeros(11)
entropy_suggestion_8am_10 = np.zeros(11)
```

```python
# ファイルのパス指定
veh_data = '../../output/{}/{}/change_num_{}/threshold_{}/veh_data.csv'.format(method, schedule, change_pseudo_num, threshold)
pseudo_data = '../../output/{}/{}/change_num_{}/threshold_{}/pseud_change.csv'.format(method, schedule, change_pseudo_num, threshold)
print(os.path.exists(veh_data))
print(os.path.exists(pseudo_data))

# csvファイルの読み込み
veh_df = pd.read_csv(veh_data)
pseud_df = pd.read_csv(pseudo_data)

pseud_df.head()

all_veh = [veh_id for veh_id in set(veh_df.id.values) if 'virtual' not in veh_id.split('_')]
all_veh_count = len(all_veh)

# 匿名セット，エントロピーの計算
eval_anonymity = dict()
for key in all_veh:
    eval_anonymity[key] = 1

for veh_id in all_veh:
    pseud_veh_id_df = pseud_df[pseud_df['id'] == veh_id]
    if not pseud_veh_id_df.empty:
        for column, item in pseud_veh_id_df.iterrows():
            around_veh_df = pseud_df[(pseud_df['candidate_point'] == item.candidate_point) \
                           & (pseud_df['step'] == item.step)]
            if len(around_veh_df.id.values) >= threshold:
                eval_anonymity[veh_id] += len(around_veh_df.id.values) - 1

mean_anonymity = np.mean(list(eval_anonymity.values()))
entropy = 0.0
for veh_id, anonymity in eval_anonymity.items():
    entropy += -1 * (1 / anonymity) * np.log2(1 / anonymity)
mean_entropy = entropy / all_veh_count
print(mean_anonymity, mean_entropy)
exec("anonymity_{}_{}_{}".format(method, schedule, threshold) + "[change_pseudo_num] = mean_anonymity")
exec("entropy_{}_{}_{}".format(method, schedule, threshold) + "[change_pseudo_num] = mean_entropy")
```
