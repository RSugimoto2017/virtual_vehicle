### SUMOのシミュレーションstepを調べてみる

```python
# 必要ライブラリをインポート
import csv, os
import numpy
import pandas as pd
```

```python
# 対象のファイルのパス文字列を指定
file_path = '../../output/3am/dua-actuated.csv'
print('file name:'.format(file_path))
```

```python
# csvからdfに変換
df = pd.read_csv(file_path)
```

```python
# 読み込んだデータの表示
df.head()
```

```python
# step数が12のものにフィルタリング
print(df[df['step'] == 12])
```

```python
# idが3BEtoDE.0にフィルタリング
# step数が1になっていれば問題なし
# departTimeが0のため
print(df[df['id'] == '3BEtoDE.0'])
```

step数が1から始まっていることがわかるため，問題なし

```python
# idがrandUni15706:1にフィルタリング
# step数が600以下からスタートしていれば問題なし
print(df[df['id'] == 'randUni15706:1'])
```

step数が587から始まっていることがわかるため，問題なし


```python
# idがrandUni15706:1にフィルタリング
# routeに関しては，初期値をぶっ込んでいるだけ
print(df[df['id'] == 'randUni15706:1'])
```

routeは残りのルートを取ってこれるようにプログラムを書き換えないといけない

```python
# ルートのindexが多い場合
print(df[df['route_index'] > 5])
```
